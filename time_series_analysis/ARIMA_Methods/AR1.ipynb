```python
import numpy as np
from matplotlib import pyplot as plt

# Basic autoregressive series of order 1, with correlation coefficient given by phi

phi = .99
sigma_w = .01

x_0 = 0
n = 1000

x = np.zeros(n)
x[0] = x_0
w = sigma_w*np.random.randn(n)
for i in range(1,n):
    x[i] = phi*x[i-1] + w[i]
    
plt.plot(range(n), x)
plt.show()
```
